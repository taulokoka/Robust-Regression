```python
import numpy as np
import matplotlib.pyplot as plt
from mm_estimator import MMEstimator
```

```python
#Generate Random data
np.random.seed(0)
A = np.random.rand(100,5)
b = np.random.rand(5)
y = A@b
plt.plot(y)
```

```python
# add outliers at random locations
y_outliers = y.copy()
y_outliers[np.random.choice(np.arange(100),10,False)] = np.random.rand(10)*1000
plt.plot(y_outliers)
```

```python
# Estimate regression coefficients 
mmest = MMEstimator()
mmest.fit(A,y_outliers)
b_hat = mmest.coef_
print('MSE coefs: ' ,np.mean((b-b_hat)**2))
print('MSE reconstruction: ' ,np.mean((y-A@b_hat)**2))

plt.plot(y)
plt.plot(A@b_hat)
```
